```python
# fatch the PDf from url
import requests
url = "https://dspmuranchi.ac.in/pdf/Blog/Python%20Built-In%20Functions.pdf"
response = requests.get(url)
```

```python
# Save the PDF to a local file
with open("Python_Built-In_Functions.pdf", "wb") as f:
    f.write(response.content)
```

```python
# loadind the document
!pip install unstructured
loader = UnstructuredFileLoader("Python_Built-In_Functions.pdf")
```

```python
!pip install "unstructured[PDF]"
```

```python
documents = loader.load()
documents
```

```python
text_splitter = CharacterTextSplitter(chunk_size=2000, chunk_overlap=400)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
type(texts)
```

```python
len(texts)
```

```python
texts[4]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = 'vector_db'
```

```python
vector_db = Chroma.from_documents(documents=texts, embedding=embeddings, persist_directory=persist_directory)
```

```python
retriever = vector_db.as_retriever()
```

```python
#LLm from groq
os.environ['GROQ_API_KEY'] = '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'
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.1)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, return_source_documents=True)
```

```python
query = "what are the functions of this PDF"
response = qa_chain.invoke({"query": query})
```
