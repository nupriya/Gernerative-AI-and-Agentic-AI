### Adding your Hugging Face Token to Colab Secrets

1.  **Get your HF Token**: Go to [Hugging Face settings](https://huggingface.co/settings/tokens) and generate a new access token (if you don't have one).
2.  **Open Colab Secrets**: In your Colab notebook, click on the "🔑" (key) icon in the left sidebar.
3.  **Add Secret**: Click 'Add new secret'.
4.  **Name the Secret**: For the 'Name' field, type `HF_TOKEN` (this is case-sensitive and important).
5.  **Paste your Token**: For the 'Value' field, paste your Hugging Face token.
6.  **Save**: Ensure 'Notebook access' is enabled for this secret.

```python
import torch
from diffusers import DiffusionPipeline
import gradio as gr
# Load Stable Diffusion XL model
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)
pipe.to("cuda")
# Image generation function
def generate_image(prompt):
    image = pipe(prompt=prompt).images[0]
    return image
# Gradio Interface
demo = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter your prompt"),
    outputs=gr.Image(label="Generated Image"),
    title="Stable Diffusion XL Image Generator",
    description="Generate AI images using Stable Diffusion XL"
)
demo.launch()
```
